## Exercício 25
Ordenar lista.

```python
lista = [1, 3, 2, 3, 1, 5, 2, 9, 7]

def bubble_sort(lista):
  for i in range(len(lista)-1):
    for j in range(len(lista)-i-1):
      if lista[j] > lista[j+1]:
        lista[j], lista[j+1] = lista[j+1], lista[j]
  return lista
bubble_sort(lista)
```

```python
def soma(lista):
  soma = 0
  for i in lista:
    soma += i
  return soma
```

## Exercício 93
Criar função que multiplica todos valores da lista.

```python
def mult(lista):
  soma = 1
  for i in lista:
    soma *= i
  return soma
```

## Exercício 94
Criar função que retorna lista invertida.

```python
def inverso(lista):
  lista = lista[::-1]
  return lista
inverso(lista)
```

## Exercício 95
Criar função que verifica se lista está ordenada.

```python
lista = [2,3,4,5,3,-2]
def ordenada(lista):
  for i in range(len(lista)-1):
    if lista[i] > lista[i+1]:
      return 1
  return 0
ordenada(lista)
```

## Exercício 96
Criar função que remove valores negativos.

```python
def remover_neg(lista):
  lista = [x for x in lista if x > 0]
  return lista
remover_neg(lista)
```

## Exercício 97
Criar função que junta duas listas.

```python
a = [1,2]
b = [3,4]
def juntar(lista1, lista2):
  lista = lista1 + lista2
  return lista
juntar(a, b)
```

## Exercício 98
Criar função que retorna interseção de listas.

```python
lista1 = [1,2,3]
lista2 = [2,3,1]
lista = [x for x in lista1 if x in lista2]
print(lista)
```

## Exercício 99
Criar função que retorna diferença entre listas.

```python
lista1 = [1,2,3]
lista2 = [2,3,1]
def dif_lista(lista1, lista2):
  lista = [x for x in lista1 if x not in lista2]
  return lista
dif_lista(lista1, lista2)
```

## Exercício 100
Criar função que conta elementos únicos.

```python
lista = [1,2,3,4,5,6,6]
unicos = []
[unicos.append(x) for x in lista if x not in unicos]
print(unicos)
```
